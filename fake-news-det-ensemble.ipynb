```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomTreesEmbedding
from sklearn.decomposition import TruncatedSVD
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('../input/textdb3/fake_or_real_news.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
X = df['text']
Y = df['label'].map(lambda x: 1 if x == 'FAKE' else 0)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=49)
```

```python
vectorizer = CountVectorizer(stop_words='english')
X_train_vectorized = vectorizer.fit_transform(X_train.values)
X_test_vectorized = vectorizer.transform(X_test.values)
```

```python
# use RandomTreesEmbedding to transform data
# hasher = RandomTreesEmbedding(n_estimators=10, random_state=0, max_depth=3)
# X_transformed = hasher.fit_transform(X_train_vectorized)

# Visualize result after dimensionality reduction using truncated SVD
#svd = TruncatedSVD(n_components=2)
#X_reduced = svd.fit_transform(X_transformed)

# Learn a Naive Bayes classifier on the transformed data
#nb = BernoulliNB()
#nb.fit(X_transformed, Y)
#print(nb)
```

```python
model = MultinomialNB(alpha=0.1)

model.fit(X_train_vectorized, y_train)

pred = model.predict(X_test_vectorized)
```

```python
# DECISION TREE CLASSIFIER
dt1 = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=2,
            max_features=None, max_leaf_nodes=40, min_samples_leaf=5,
            min_samples_split=2, min_weight_fraction_leaf=0.0,
            random_state=None, splitter='random')
# train the model
dt1.fit(X_train_vectorized, y_train)
# make predictions
dt_pred1 = dt1.predict(X_test_vectorized)
```

```python
print(accuracy_score(y_test, pred))
print(accuracy_score(y_test, dt_pred1))
```

```python
print(confusion_matrix(y_test, pred))
print(confusion_matrix(y_test, dt_pred1))
```

```python
from sklearn.linear_model import LogisticRegression #logistic regression
from sklearn import svm #support vector Machine
from sklearn.ensemble import RandomForestClassifier #Random Forest
from sklearn.neighbors import KNeighborsClassifier #KNN
from sklearn.naive_bayes import GaussianNB #Naive bayes
from sklearn.tree import DecisionTreeClassifier #Decision Tree
from sklearn.model_selection import train_test_split #training and testing data split
from sklearn import metrics #accuracy measure
from sklearn.metrics import confusion_matrix #for confusion matrix
from sklearn.ensemble import VotingClassifier

ensemble1=VotingClassifier(estimators=[('LR',LogisticRegression(C=0.05)),
                                              ('DT',DecisionTreeClassifier(random_state=0)),
                                              ('NB',GaussianNB())], 
                       voting='soft').fit(X_train_vectorized.todense(),y_train)
print('The accuracy for ensembled model 1 is:',ensemble1.score(X_test_vectorized.todense(),y_test))


# Ensemble 2

ensemble2=VotingClassifier(estimators=[('LR',LogisticRegression(C=0.1)),
                                              ('DT',DecisionTreeClassifier(random_state=0)),
                                              ('NB',GaussianNB())], 
                       voting='soft').fit(X_train_vectorized.todense(),y_train)
print('The accuracy for ensembled model 2 is:',ensemble2.score(X_test_vectorized.todense(),y_test))

# Ensemble 3

ensemble3=VotingClassifier(estimators=[('LR',LogisticRegression(C=1)),
                                              ('DT',DecisionTreeClassifier(random_state=0)),
                                              ('NB',GaussianNB())], 
                       voting='soft').fit(X_train_vectorized.todense(),y_train)
print('The accuracy for ensembled model 3 is:',ensemble3.score(X_test_vectorized.todense(),y_test))

# Ensemble 4

ensemble4=VotingClassifier(estimators=[('LR',LogisticRegression(C=0.01)),
                                              ('DT',DecisionTreeClassifier(random_state=0)),
                                              ('NB',GaussianNB())], 
                       voting='soft').fit(X_train_vectorized.todense(),y_train)
print('The accuracy for ensembled model 4 is:',ensemble4.score(X_test_vectorized.todense(),y_test))
```
